```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Titanic-Dataset.csv")
print(df.head())
print(df.info())
print(df.describe(include='all'))
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df['Cabin'] = df['Cabin'].fillna("Unknown")
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = 1
df.loc[df['FamilySize'] > 1, 'IsAlone'] = 0
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title("Survival Count")
plt.show()
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set2')
plt.title("Survival by Gender")
plt.show()
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set2')
plt.title("Survival by Passenger Class")
plt.show()
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.show()
sns.boxplot(x='Survived', y='Age', data=df, palette='Set2')
plt.title("Age vs Survival")
plt.show()
sns.histplot(df['Fare'], bins=20, kde=True, color='lightgreen')
plt.title("Fare Distribution")
plt.show()
sns.countplot(x='FamilySize', hue='Survived', data=df, palette='Set2')
plt.title("Family Size vs Survival")
plt.show()
sns.countplot(x='Embarked', hue='Survived', data=df, palette='Set2')
plt.title("Survival by Embarkation Port")
plt.show()


```
